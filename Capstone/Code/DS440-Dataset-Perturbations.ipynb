### Row Uniformity
Take out a percentage of the data and replace the removed rows with random duplicates

```python
def add_duplicate_rows(df, p=.25, reset_index=False):
  df2 = pd.concat([df, df.sample(n=int(len(df)*p))])

  if reset_index: 
    return df2.reset_index(drop=True)
  else: 
    return df2
```

```python
duplicate_rows  = add_duplicate_rows(new_df, p=.25)
duplicate_rows.to_csv('2012_2013_duplicate_rows.csv')
```

### Change the data
Add noise


```python
def add_col_noise(df, ignore_cols=['TeamId', 'GameId','Date']):
  for col in df.columns.to_list():
    if col in ignore_cols:
      continue
    else:
      std = df[col].std()
      noise = np.random.normal(0, std, len(df[col]))
      df[col] += noise
  return df


add_col_noise_df  = add_col_noise(new_df)
add_col_noise_df.to_csv('2012_2013_add_noise.csv')
```

### Feature Statistics: Create synthetic data
- Using the statistics of each feature: mean, variance, min, max, etc; or use another probabilistic model.
- Using changed statistics of each feature: mean, variance, min, max, etc; or use another probabilistic model.


```python
def bootstrapped_df(df, keep_cols=['TeamId', 'GameId','Date']):
  df2 = pd.DataFrame()
  for col in df.columns.to_list():
    if col in keep_cols:
      df2[col] = df[col]
    else:
      mu = df[col].mean()
      std = df[col].std()
      noise = np.random.normal(mu, std**.5, len(df[col]))
      df2[col] = noise

  return df2

bootstrapped_dataframe  = bootstrapped_df(new_df)
bootstrapped_dataframe.to_csv('2012_2013_bootstrapped.csv')
```

### Drop features




```python
def drop_features(df, n_cols=2, keep_columns=['TeamId', 'GameId','Date']):
  columns = df.drop(keep_columns, axis=1).columns.to_list()
  drop = np.random.choice(columns, size=n_cols)
  return df.drop(drop, axis=1)

#drop_features(teams_odds_moneyline)
drop_features_dataframe  = drop_features(new_df)
drop_features_dataframe.to_csv('2012_2013_drop_features.csv')
```

### Add dummy features

```python
def add_dummy_cols(df, n_cols=2):
  for i in range(n_cols):
    df[i + 10000] = np.random.normal(size=len(df))
  return df

#add_dummy_cols(teams_odds_moneyline)
add_dummy_cols_dataframe  = add_dummy_cols(new_df)
add_dummy_cols_dataframe.to_csv('2012_2013_add_dummy_cols.csv')
```
